```python
df.head(10)
```

```python
dup = df[df['ID'].duplicated()]
df.drop_duplicates(subset ="ID", keep = False, inplace = True)
```

```python
df.columns
```

```python
df['pe'][df['pe'] == 'Left'] = 'Esquerdo'
df['pe'][df['pe'] == 'Right'] = 'Direito'
```

```python
df['dedicacao_ataque'][df['dedicacao_ataque'] == 'Low'] = 'Baixa'
df['dedicacao_defesa'][df['dedicacao_defesa'] == 'Low'] = 'Baixa'
df['dedicacao_ataque'][df['dedicacao_ataque'] == 'Medium'] = 'Média'
df['dedicacao_defesa'][df['dedicacao_defesa'] == 'Medium'] = 'Média'
df['dedicacao_ataque'][df['dedicacao_ataque'] == 'High'] = 'Alta'
df['dedicacao_defesa'][df['dedicacao_defesa'] == 'High'] = 'Alta'
```

```python
df['selecao'][df['selecao'] == 'Not in team'] = 'Fora'
```
