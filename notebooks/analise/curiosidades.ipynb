### Jogadores por partes do campo

```python
positions = ['GK', 'CB', 'LB', 'RB', 'LWB', 'RWB', 'CDM', 'CM', 'LM', 'RM', 'CAM', 'LW', 'RW', 'CF', 'ST']
values = []
defesa_count = meio_count = ataque_count = 0
defesa_pct = meio_pct = ataque_pct = 0

for i in positions:
    values.append(len(df[df.posicao_preferida == i]))

defesa = values[0:6]
meio = values[6:11]
ataque = values[11:15]

for d in defesa:
    defesa_count += d
    defesa_pct = np.round((defesa_count*100) / 19248, 2)
    
for m in meio:
    meio_count += m
    meio_pct = np.round((meio_count*100) / 19248, 2)
    
for a in ataque:
    ataque_count += a
    ataque_pct = np.round((ataque_count*100) / 19248, 2)

print('Defesa: ' + str(defesa_pct) + '%\n' + str(defesa_count) + '\n')
print('Meio:   ' + str(meio_pct) + '%\n' + str(meio_count) + '\n')
print('Ataque: ' + str(ataque_pct) + '%\n' + str(ataque_count))

plt.title('FIFA 22 - Gráfico de Porcentagem de jogadores por parte de campo (Barras)')
plt.bar(['Defesa', 'Meio', 'Ataque'], [defesa_pct, meio_pct, ataque_pct], width=0.3)
plt.show()
```

```python
def radar(titulo, valores, atributos):
    
    plt.rcParams.update({
        "figure.facecolor":  '#fff',
        "axes.facecolor":    '#fff',
    })

    
    nValores = []
    for i in valores:
        nValores.append(i-70)

    nValores.pop(0)
    atributos.pop(0)
    
    ##Criando novo array com valores menore para melhor apresentação gráfica e removido atributo overall
        
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    theta = np.arange(len(atributos) + 1) / float(len(atributos)) * 2 * np.pi
    values = np.append(nValores, nValores[0])
    
    l1, = ax.plot(theta, values, color="blue", marker="o", label="Valores")
    plt.xticks(theta[:-1], atributos, color='#000', size=10)
    ax.tick_params(pad=10)
    ax.fill(theta, values, 'blue', alpha=0.1)
    ax.plot(linewidth=1, linestyle='solid', edgecolor='white')

    ax.set_yticklabels([])
    
    plt.title(titulo, pad=25)
    return plt.show()
```

```python
def radar(titulo, valores, atributos):
    nValores = []
    for i in valores:
        nValores.append(i-70)
            
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    theta = np.arange(len(atributos) + 1) / float(len(atributos)) * 2 * np.pi
    values = np.append(nValores, nValores[0])
    
    l1, = ax.plot(theta, values, color="blue", marker="o", label="Valores")
    plt.xticks(theta[:-1], atributos, color='#000', size=10)
    ax.tick_params(pad=10)
    ax.fill(theta, values, 'blue', alpha=0.1)
    ax.plot(linewidth=1, linestyle='solid', edgecolor='white')

    ax.set_yticklabels([])
    
    plt.title(titulo, pad=25)
    return plt.show()
```
